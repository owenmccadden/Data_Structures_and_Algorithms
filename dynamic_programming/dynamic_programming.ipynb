# Dynamic Programming
- algorithmic design technique for optimization problems
- break global problem down into simpler subproblems, utilize the fact that the optimal global solution depends on the optimal solution to its subproblems
- solve subproblem once and reuse previous solution
- a more optimized divide and conquer approach
    - less repition of already solved problems
- optimal substructure is good use case for dynamic programming
- overlapping subproblem: smaller versions of original problem
    - finding solution involves solving same sub problem many times

## Top Down: Memoization
- solve bigger problem by recursively finding solution to smaller subproblems
- cache the result of each subproblem so that we don't have to solve it repeatedly if it gets called again
- converting a divide and conquer solution to a dynamic programming top down solution:
    - add a cache array, dictionary, etc. to store solutions to solved problems as a parameter to method
    - check if solution exists in cache for each call
    - update cahce with new value on each method call
    - return value from cache


## Bottom Up: Tabulation
- opposite of memoization, avoids recursion
- solve problem bottom up –– solve subproblem first
- fill up a table with the results of the sub problems
- compute the original problem from the table
