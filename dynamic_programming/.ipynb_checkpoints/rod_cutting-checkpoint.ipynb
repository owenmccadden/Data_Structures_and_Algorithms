# Rod Cutting

- given a list of integers representing the value of a rod of length i+1, return a list of the maximum revenue attainable for a pipe size of i

### Example:
[2, 5, 6, 10, 11, 15, 16, 28]



[5, 6, 7, 10, 11, 12, 14, 15]



```python
def bottom_up(p):
    n = len(p)
    r = [0]
    for j in range(0,n):
        q = -math.inf
        for i in range(0,j+1):
            q = max(q, p[i]+r[j-i])
        r.append(q)
    return r[1:]
```
