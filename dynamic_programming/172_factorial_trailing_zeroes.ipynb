```python
class Solution:
    def trailingZeroes(self, n: int) -> int:
        value = self.factorial(n)
        count = 0
        for digit in str(value)[::-1]:
            if digit == '0':
                count += 1
            else:
                break
        return count
    
    def factorial(self, n):
        cache = [1]
        for i in range(1, n+1):
            cache.append(i * cache[i-1])
        return cache[n]
```
