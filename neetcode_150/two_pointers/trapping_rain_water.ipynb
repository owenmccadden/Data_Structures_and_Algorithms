return the amount of rainwater that could be collected given a list of bar heights



```python
"""

brute force solution O(n) time,  O(n) space

"""


def trap(heights):
    area = []
    max_left = []
    max_right = []
    
    for i, height in enumerate(heights):
        if i == 0:
            max_left.append(0)
        else:
            max_left.append(max(heights[:i]))
            
        if i == len(heights) - 1:
            max_right.append(0)
        else:
            max_right.append(max(heights[i+1:]))
            
        area.append(max(min(max_left[i], max_right[i]) - height, 0))
        
    return sum(area)
```

```python
heights = [0,1,0,2,1,0,1,3,2,1,2,1]
```

```python
trap(heights)
```

```python
"""

two pointers solution

"""

def trap(height):
    left = 0
    right = len(height) - 1
    
    max_left = height[left]
    max_right = height[right]
    
    area = 0
    
    while left < right:
        if max_left < max_right:
            left += 1
            max_left = max(max_left, height[left])
            area += max_left - height[left]
        else:
            right -= 1
            max_right = max(max_right, height[right])
            area += max_right - height[right]
    
    return area
```

```python
trap(heights)
```
