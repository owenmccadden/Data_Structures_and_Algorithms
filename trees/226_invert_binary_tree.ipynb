# 226_invert_binary_tree


```python
from collections import deque
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def levelorder(head):
    q = deque()
    visited = []
    q.appendleft(head)
    while q:
        curr = q.pop()
        visited.append(curr.val)
        if curr.left is not None:
            q.appendleft(curr.left)
        if curr.right is not None:
            q.appendleft(curr.right)
    return visited
```

```python
def invert_tree(head):
    if head is None:
        return head
    left = invert_tree(head.left)
    right = invert_tree(head.right)
    
    head.left = right
    head.right = left
    
    return head
```

```python
test1 = TreeNode(4, TreeNode(2, TreeNode(1), TreeNode(3)), TreeNode(7, TreeNode(6), TreeNode(9)))
levelorder(test1)
```

```python
result = invert_tree(test1)
levelorder(result)
```
