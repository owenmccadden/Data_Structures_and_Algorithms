# 114 Flatten Binary Tree to Linked List
- given a binary tree, flatten it to a linked list in place
- remove all left nodes and put them into a list along the right node path, should follow same order as pre-order traversal


```python
from collections import deque
```

```python
class TreeNode:
    def __init__(self, val, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
    
    def insert_left(self, data):
        self.left = TreeNode(data)
        return self.left
    
    def insert_right(self, data):
        self.right = TreeNode(data)
        return self.right

    def preorder(self):
        print(self.val)
        if self.left:
            self.left.preorder()
        if self.right:
            self.right.preorder()
    
    def traverse_right(self):
        print(self.val)
        if self.right:
            self.right.traverse_right()
```

```python
example = TreeNode(1)
example.insert_left(2)
example.insert_right(5)
example.left.insert_left(3)
example.left.insert_right(4)
example.right.insert_right(6)
```

```python
example.traverse_right()
```

```python
def to_linked_list(root):
    stack = deque()
    stack.append(root)
    while stack:
        curr = stack.pop()
        if  curr.right:
            stack.append(curr.right)
        if curr.left:
            stack.append(curr.left)
        if stack:
            curr.right = stack[-1]
        curr.left = None
```

```python
to_linked_list(example)
```

```python
example.traverse_right()
```
