```python
from collections import deque
```

```python
class Node:
    def __init__(self, val=None, children=None):
        self.val=val
        self.children=children
```

```python
test = Node(1, [Node(3, [Node(5, []), Node(6, [])]), Node(2, []), Node(4, [])])
```

```python
def levelorder_traversal(head):
    q = deque()
    visited = []
    q.appendleft(head)
    while len(q) > 0:
        curr = q.pop()
        visited.append(curr.val)
        for i in curr.children:
            q.appendleft(i)
    return visited
```

```python
levelorder_traversal(test)
```

```python
def preorder_traversal(head):
    stack = deque()
    visited = []
    stack.append(head)
    while len(stack) > 0:
        curr = stack.pop()
        visited.append(curr.val)
        for i in curr.children[::-1]:
            stack.append(i)
    return visited
```

```python
preorder_traversal(test)
```

```python
def postorder_traversal(head):
    stack = deque()
    visited = deque()
    stack.append(head)
    while len(stack) > 0:
        curr = stack.pop()
        visited.appendleft(curr.val)
        for i in curr.children:
            stack.append(i)
    return visited
```

```python
postorder_traversal(test)
```
