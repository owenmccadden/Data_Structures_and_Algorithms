# 112 Path Sum
- given a treenode and a target sum, return true if the tree has a root to leaf path that equals target sum


```python
from collections import deque
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val=val
        self.left=left
        self.right=right
```

```python
def levelorder(root):
    q = deque()
    visited = []
    q.appendleft(root)
    while q:
        curr = q.pop()
        visited.append(curr.val)
        if curr.left is not None:
            q.appendleft(curr.left)
        if curr.right is not None:
            q.appendleft(curr.right)
    return visited
```

```python
test1 = TreeNode(5, TreeNode(4, TreeNode(11, TreeNode(7), TreeNode(2))), TreeNode(8, TreeNode(13), TreeNode(4, None, TreeNode(1))))
```

```python
levelorder(test1)
```

```python
def path_sum(root, target):
    if root is None:
        return False
    
    node_stack = deque()
    sum_stack = deque()
    
    node_stack.append(root)
    sum_stack.append(target-root.val)
    
    while node_stack:
        curr_node = node_stack.pop()
        curr_sum = sum_stack.pop()
        
        if curr_node.left is None and curr_node.right is None and curr_sum == 0:
            return True
        
        if curr_node.left is not None:
            node_stack.append(curr_node.left)
            sum_stack.append(curr_sum-curr_node.left.val)
            
        if curr_node.right is not None:
            node_stack.append(curr_node.right)
            sum_stack.append(curr_sum-curr_node.right.val)
            
    return False
```

```python
path_sum(test1, 22)
```
