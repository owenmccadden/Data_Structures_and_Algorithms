# 103: Zigzag Level Order Traversal


```python
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def insert_left(self, data):
        self.left = BinaryTreeNode(data)
        return self.left
    
    def insert_right(self, data):
        self.right = BinaryTreeNode(data)
        return self.right
    
    def bfs(self):
        queue = [self]
        while len(queue) > 0:
            current = queue.pop(0)
            print(current.data)
            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)
```

```python
test = BinaryTreeNode(3)
test.left = BinaryTreeNode(9)
test.right = BinaryTreeNode(20)
test.right.left = BinaryTreeNode(15)
test.right.right = BinaryTreeNode(7)
test.bfs()
```

```python
class Solution:
    def zigzag_level_order(self, root):
        self.rows = []
        self.fill_rows(root, 0)
        for i in range(len(self.rows)):
            if i % 2 != 0:
                self.rows[i] = self.rows[i][::-1]
        return self.rows
        
    def fill_rows(self, root, level):
        if root is None:
            return
        if len(self.rows) <= level:
            self.rows.append([])
        self.rows[level].append(root.data)
        self.fill_rows(root.left, level+1)
        self.fill_rows(root.right, level+1)
```

```python
test1 = Solution()
test1.zigzag_level_order(test)
```
