# 617_merge_two_binary_trees
- given two binary trees, return the result of merging the trees
- when the trees overlap, set the value to the sum of the overlapping nodes


```python
from collections import deque
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def levelorder(head):
    q = deque()
    visited = []
    q.appendleft(head)
    while q:
        curr = q.pop()
        visited.append(curr.val)
        if curr.left is not None:
            q.appendleft(curr.left)
        if curr.right is not None:
            q.appendleft(curr.right)
    return visited
```

```python
def merge_tree(root1, root2):
    if root1 is None:
        return root2
    elif root2 is None:
        return root1
    else:
        root1.val += root2.val
    root1.left = merge_tree(root1.left, root2.left)
    root1.right = merge_tree(root1.right, root2.right)
    return root1;
```

```python
test1 = TreeNode(1, TreeNode(3, TreeNode(5)), TreeNode(2))
levelorder(test1)
```

```python
test2 = TreeNode(2, TreeNode(1, None, TreeNode(4)), TreeNode(3, None, TreeNode(7)))
levelorder(test2)
```

```python
result = merge_tree(test1, test2)
levelorder(result)
```
