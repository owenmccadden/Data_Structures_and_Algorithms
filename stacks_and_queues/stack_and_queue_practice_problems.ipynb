# Stack and Queue Implementations using Deque

```python
from collections import deque
```

```python
class Stack:
    def __init__(self, values=[]):
        self.list = deque(values)
    
    def push(self, value):
        self.list.append(value)
    
    def pop(self):
        if self.is_empty():
            return
        else:
            return self.list.pop()
    
    def peek(self):
        return self.list[-1]
    
    def is_empty(self):
        if len(self.list) == 0:
            return True
        else:
            return False
    
    def size(self):
        return len(self.list)
```

```python
class Queue:
    def __init__(self, values=[]):
        self.list = deque(values)
    
    def is_empty(self):
        if len(self.list) == 0:
            return True
        else:
            return False
    
    def enqueue(self, value):
        self.list.appendleft(value)
    
    def dequeue(self):
        if self.is_empty():
            return None
        else:
            return self.list.pop()
    
    def peek(self):
        return self.list[-1]
    
    def size(self):
        return len(self.list)
```

# Question 1:
- use a single python list or deque to implement three stacks of equal length

```python
class tripleStack:
    def __init__(self, stack_size):
        self.stack_size = stack_size
        self.num_stacks = 3
        self.list = [0] * self.num_stacks * self.stack_size
        self.sizes  = [0] * self.num_stacks
    
    def __str__(self):
        return ''.join(str(self.list))
    
    def is_full(self, stack_num):
        if self.sizes[stack_num] == self.stack_size:
            return True
        else:
            return False
    
    def is_empty(self, stack_num):
        if self.sizes[stack_num] == 0:
            return True
        else:
            return False
    
    def index_of_top(self, stack_num):
        offset = stack_num * self.stack_size
        return offset + self.sizes[stack_num]
    
    def push(self, value, stack_num):
        if self.is_full(stack_num):
            print('Stack {} is full.'.format(stack_num))
            return
        else:
            self.list[self.index_of_top(stack_num)] = value
            self.sizes[stack_num] += 1
    
    def pop(self, stack_num):
        if self.is_empty(stack_num):
            print('Stack {} is empty.'.format(stack_num))
            return None
        else:
            value = self.list[self.index_of_top(stack_num)]
            self.list[self.index_of_top(stack_num)] = 0
            self.sizes[stack_num] -= 1
            return value
    
    def peek(self, stack_num):
        if self.is_empty(stack_num):
            return None
        else:
            return self.list[index_of_top(stack_num)]
            
    
        
    
```

        - problem is in this case, if the value being popped is the min, we have to find a way to reassign the min in constant time
        - ...
        
        
- linked list implementation solves this issue and ensure that min has constant time complexity by keeping a link from min to the previous min

```python
class MinStack:
    def __init__(self, values=[]):
        self.list = deque(values)
        self.min = 0
    
    def __str__(self):
        return ''.join(str(self.list))
    
    def push(self, value):
        if self.is_empty() or value < self.min:
            self.min = value
        self.list.append(value)
    
    def pop(self):
        if self.is_empty():
            return
        else:
            value = self.peek()
            if value == self.min:
                self.list.pop()
                self.min = min(self.list)
                return value
            else:
                return self.list.pop()
    
    def min_value(self):
        return self.min
    
    def peek(self):
        return self.list[-1]
    
    def is_empty(self):
        if len(self.list) == 0:
            return True
        else:
            return False
    
    def size(self):
        return len(self.list)
```

# Question 3:
- implement a data structure set of stacks that is composed of several stacks and creates a new stack once previous one exceeds capacity
- push and pop should operate like a regular stack i.e. operate only on a particular stack in the set
- include a method popAt which performs a pop operation on a specified sub stack

```python
class SetOfStacks:
    def __init__(self, num_stacks, stack_size):
        self.num_stacks = num_stacks
        self.stack_size = stack_size
        self.stacks = dict()
        self.current_stack = 0
        for i in range(num_stacks):
            self.stacks[i] = []
    
    def is_full(self, stack_num):
        if len(self.stacks[stack_num]) == self.stack_size:
            return True
        else:
            return False
    
    def is_empty(self, stack_num):
        if len(self.stacks[stack_num]) == 0:
            return True
        else:
            return False
    
    def push(self, value):
        if self.is_full(self.current_stack):
            if self.current_stack == self.num_stacks - 1:
                print('Set of stacks is full.')
                return
            else:
                self.current_stack += 1
                self.push(value)
        else:
            self.stacks[self.current_stack].append(value)
    
    def pop(self, stack_num):
        if self.is_empty(stack_num):
            print('stack {} is empty.'.format(stack_num))
            return None
        else:
            return self.stacks[stack_num].pop()
    
    def peek(self, stack_num):
        if self.is_empty(stack_num):
            print('stack {} is empty.'.format(stack_num))
            return None
        else:
            return self.stacks[stack_num][-1]
```

```python
a = SetOfStacks(5, 2)
for i in range(6):
    a.push(1)
    a.push(2)
    print(a.stacks)
a.pop(2)
a.stacks
```

# Question 5:
- animal shelter holds only dogs and cats
- people can either adopt the animal that has been there the longest, or specify a preference for dog or cat and then adopt the oldest of animal of their preference
- design and implement a data structure to maintain this system
    - use 'cat' to represent a cat and 'dog' to represent a dog

```python
class AnimalShelter:
    def __init__(self):
        self.animals = deque()
        self.cat_count = 0
        self.dog_count = 0
    
    def is_empty(self):
        return len(self.animals) == 0
    
    def has_cats(self):
        return 'cat' in self.animals
    
    def has_dogs(self):
        return 'dog' in self.animals
    
    def enqueue(self, animal):
        self.animals.append(animal)
        if animal == 'cat':
            self.cat_count += 1
        elif animal == 'dog':
            self.dog_count += 1
        else:
            raise ValueError("The shelter does not accept {}".format(animal))
    
    def dequeue(self, preference):
        if self.is_empty():
            print("The shelter is empty.")
            return
        
        if preference == 'cat':
            if self.cat_count == 0:
                print('No cats remaining.')
                return
            for i in range(len(self.animals)):
                if self.animals[i] == 'cat':
                    pet = self.animals[i]
                    self.animals.remove('cat')
                    return pet
        elif preference == 'dog':
            if self.dog_count == 0:
                print('No dogs remaining.')
                return
            for i in range(len(self.animals)):
                if self.animals[i] == 'dog':
                    pet = self.animals[i]
                    self.animals.remove('dog')
                    return pet
        else:
            return self.animals.popleft()
    
    def peek(self):
        if self.is_empty():
            return
        else:
            return self.animals[0]
```
