# 141 Linked List Cycle
- return a boolean to indicate if a linked list contains a cycle


```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def has_cycle(head):
    if head is None:
            return False
    visited = set()
    while head is not None:
        if head in visited:
            return True
        else:
            visited.add(head)
            head = head.next
        
    return False
```

```python
three = ListNode(3)
two = ListNode(2)
zero = ListNode(0)
four = ListNode(-4)
three.next = two
two.next = zero
zero.next = four
four.next = two
```

```python
has_cycle(three)
```

```python
test2 = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))
```

```python
has_cycle(test2)
```

```python
def has_cycle(head):
    if head is None:
        return False
    
    slow = head
    fast = head.next
    
    while slow != fast:
        if fast is None or fast.next is None:
            return False
        
        slow = slow.next
        fast = fast.next.next
    
    return True
```

```python
has_cycle(three)
```

```python
has_cycle(test2)
```
