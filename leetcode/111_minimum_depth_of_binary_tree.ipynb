# 104 Minimum Depth of Binary Tree
- return the number of nodes along the longest path from the root to the farthest tree

- in this ex, max depth = 2

```python
from collections import deque
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def min_depth(root, cache):
    if root is None:
        return 0
    if root.left is None:
        cache[root] = min_depth(root.right, cache) + 1
    if root.right is None:
        cache[root] = min_depth(root.left, cache) + 1
    if root not in cache:
        cache[root] = 1 + min(min_depth(root.left, cache), min_depth(root.right, cache))
    return cache[root]
```

```python
test1 = TreeNode(3, TreeNode(9), TreeNode(20, TreeNode(15), TreeNode(7)))
```

```python
min_depth(test1, {})
```
