# 22 Generate Parentheses
- given an integer n, return a list of all valid pairs of n parentheses

```python
class Solution:
    def generateParenthesis(self, n):
        self.result = []
        if n > 0:
            self.dfs(n, '', 0)
        return self.result
    
    def dfs(self, n, temp, index):
        if index == n*2:
            if self.is_valid(temp, n):
                self.result.append(temp)
            return
        
        for ch in ('(', ')'):
            temp += ch
            self.dfs(n, temp, index+1)
            temp = temp[:-1]
            
    def is_valid(self, temp, n):
        for i in range(n):
            temp = temp.replace('()','')
        return not temp
```

```python
for i in range(7):
    print(test.generateParenthesis(i))
```
