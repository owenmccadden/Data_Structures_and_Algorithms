# 79: Word Search
- return a boolean to indicate whether a given word exists in a 2d matrix of characters
- letters must be next to each other front back left right, not diag


```python
class Solution:
    def exist(self, board, word):
        self.found_word = False
        for i in range(len(board)):
            for j in range(len(board[0])):
                if self.found_word:
                    return self.found_word
                else:
                    self.search(board, i, j, word, 0, set(), '')
        return self.found_word
    
    def search(self, board, i, j, word, index, visited, temp):
        if not board or i >= len(board) or j >= len(board[0]) or i < 0 or j < 0 or index > len(word):
            return
        if board[i][j] != word[index]:
            return
        temp += board[i][j]
        visited.add((i,j))
        if temp == word:
            self.found_word = True
            return
        if (i+1, j) not in visited:
            self.search(board, i+1, j, word, index+1, visited, temp)
        if (i-1, j) not in visited:
            self.search(board, i-1, j, word, index+1, visited, temp)
        if (i, j+1) not in visited:
            self.search(board, i, j+1, word, index+1, visited, temp)
        if (i, j-1) not in visited:
            self.search(board, i, j-1, word, index+1, visited, temp)
        visited.remove((i,j))
```
