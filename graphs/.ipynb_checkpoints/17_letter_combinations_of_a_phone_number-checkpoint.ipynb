# 17: Letter Combinations of a Phone Number
- given a string of digits (2-9), return all possible letter combinations that the number could be
- mapping of numbers to letters:


```python
class Solution:
    def letterCombinations(self, digits):
        self.result = []
        if digits:
            self.map = {
                '2': ['a', 'b', 'c'],
                '3': ['d', 'e', 'f'],
                '4': ['g', 'h', 'i'],
                '5': ['j', 'k', 'l'],
                '6': ['m', 'n', 'o'],
                '7': ['p', 'q', 'r', 's'],
                '8': ['t', 'u', 'v'],
                '9': ['w', 'x', 'y', 'z']
            }
            self.dfs(digits, self.map, self.result, '', 0)
            
        return self.result
            
    def dfs(self, digits, n_map, result, temp, index):
        if index == len(digits):
            result.append(temp)
            return
        
        digit = digits[index]
        letters = self.map[digit]
        
        for ch in letters:
            temp += ch
            self.dfs(digits, n_map, result, temp, index+1)
            temp = temp[:-1]
            
```
