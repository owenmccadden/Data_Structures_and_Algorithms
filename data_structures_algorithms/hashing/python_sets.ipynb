# Sets

## Properties
- iterable where all elements are unordered, unindexed, and unique
- cannot remove items from a set, but can add items
- cannot have duplicates
- can be of any data type or different data types


## Operations

### Initialization
- set_name = {item1, item2}
- can also use set() constructor
    - a = set((1,2,3))
    - must use set constructor to declare an empty set
- O(1) time complexity
- O(n) space complexity

```python
my_set = {1,2,3,4,5}
```

```python
my_set
```

```python
a = set((1,2,3))
a
```

### len()
- returns number of items in set

```python
len(my_set)
```

### Accessing items
- cannot access individual items using an index or key
- can iterate through items in set using for loop
- can use in operator to check if items exist in set
- cannot modify or delete items once they are added to set

```python
a = {1,2,3}
2 in a
```

```python
5 in a
```

```python
for i in a:
    print(i)
```

### adding items
- add new items using the add(value) method
- use the update(iterable) method to merge any set with another set or with any other iterable

```python
print(a)
```

```python
a.add(4)
print(a)
```

```python
a.add(1)
print(a)
```

```python
b = set()
b.add(5)
b.add(6)
b.add(7)
b
```

```python
a.update(b)
a
```

```python
c = [11,12,13,14,15]
a.update(c)
a
```

### removing items
- pop removes the last item in a set (most recently added)
- remove(val) or discard(val) removes a particular value from a set
- clear removes all values from a set

```python
print(a)
a.remove(1)
print(a)
```

```python
print(a)
a.discard(4)
print(a)
```

```python
print(a)
a.pop()
print(a)
```

```python
print(a)
a.clear()
print(a)
```

### unions and intersections
- update joins all values from each set
- intersection_update joins only values present in both sets
- symmetric_different_update keeps only elements present not in both sets

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.update(y)
print(x)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.intersection_update(y)
print(x)
```

```python
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}
x.symmetric_difference_update(y)
print(x)
```
