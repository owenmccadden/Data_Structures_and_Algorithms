# Strings
- similar to most programming languages, strings are arrays of bytes representing unicode characters
- no char data type in python, instead strings of length 1
- accessing, modifying, length, slicing, in operator, etc. are identical to python lists

# String Methods

### Upper and Lower
- return upper / lower case versions of string

```python
a = 'owen'
a.upper()
```

```python
a = "MCCADDEN"
a.lower()
```

### Strip
- removes whitespace from beginning and end of string

```python
a = '   hello   '
a.strip()
```

### Replace
- replaces a substring of a given string with another string
- replace('str', '') can be used to remove all instances of a particular str in a string

```python
a = 'a:b:c:d:f:g:G:H:R'
b = a.replace(':', '^^')
b
```

```python
b.replace('^','')
```

### Split
- returns a list of substrings that are split by a separator string
- by default the separator is ' '


```python
a = 'this is a string.'
a.split(' ')
```

### Concatentation (+ operator)
- use + to concatenate two strings

```python
a = 'this is '
b = 'a string.'
a + b
```

### Formatting
- use {} and .format to insert values into a string

```python
a = 'my name is {} and I am {} years old.'.format('Owen', 20)
a
```

```python
a = 'here is the {0} example of using the {1} method.'.format('2nd', '.format')
a
```

### Escape Characters

```python
print('here is a string\' that contains \' some\t escape characters.\n')
```

### Capitalize:
- converts first letter to uppercase

```python
a = 'owen'
a.capitalize()
```

### Casefold
- converts string to lowercase

```python
a = "OWEN"
a.casefold()
```

### Center
- returns a centered string
- returns original string with n/2 spaces to the left and right

```python
a = 'owen'
a.center(100)
```

### Encode
- returns an encoded version of string
- i.e. converts data type to bytes

```python
a = 'this is a string'
a.encode()
```

### Endswith
- returns true if string ends with specified value

```python
a = 'string'
a.endswith('g')
```

### Find
- returns the index of first instance of a specified value
- returns -1 if no value is found

```python
s = 'string string'
s.find('ring')
```
