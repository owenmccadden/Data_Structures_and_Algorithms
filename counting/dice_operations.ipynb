```python
import itertools
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plot
```

```python
# assume n is fixed at 2 for subtraction and division

def dice_ops(n, m, op):
    results = {}
    sum_counts = defaultdict(int)
    possible_rolls = defaultdict(list)
    total_combs = 0
    dice_roll = [i for i in range(1, m+1)]
    probabilities = {}
    
    for comb in itertools.product(*[[i for i in range(1, m+1)] for j in range(n)]):
        if op == 'sum':
            comb_result = sum(comb)
        elif op == 'difference':
            max_val = max(comb)
            min_val = min(comb)
            comb_result = max_val - min_val
        elif op == 'product':
            product = 1
            for num in comb:
                product *= num
            comb_result = product
        elif op == 'quotient':
            max_val = max(comb)
            min_val = min(comb)
            comb_result = max_val / min_val
        sum_counts[comb_result] += 1
        possible_rolls[comb_result].append(comb)
        total_combs += 1
        
    for key, value in sum_counts.items():
        probabilities[key] = round(value / total_combs * 100, 3)
        results[key] = (value, probabilities[key], possible_rolls[key])
    
    plot.scatter(*zip(*probabilities.items()))
    plot.show()
        
    return (sum_counts, possible_rolls, probabilities, total_combs)
```

```python
dice_ops(2, 6, 'sum')
```
