```python
import itertools
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plot
```

```python
def dice_sums(n, m):
    results = {}
    sum_counts = defaultdict(int)
    possible_rolls = defaultdict(list)
    total_combs = 0
    dice_roll = [i for i in range(1, m+1)]
    probabilities = {}
    
    for comb in itertools.product(*[[i for i in range(1, m+1)] for j in range(n)]):
        comb_sum = sum([num for num in comb])
        sum_counts[comb_sum] += 1
        possible_rolls[comb_sum].append(comb)
        total_combs += 1
        
    for key, value in sum_counts.items():
        probabilities[key] = round(value / total_combs * 100, 3)
        results[key] = (value, probabilities[key], possible_rolls[key])
    
    plot.scatter(*zip(*probabilities.items()))
    plot.show()
        
    return (sum_counts, possible_rolls, probabilities, total_combs)
```

# Two Six-Sided Die

```python
dice_sums(2, 6)
```

# All combinations of n and m in range 8

```python
combs = defaultdict(list)

for i in range(1, 9):
    for j in range(1, 9):
        result = dice_sums(j, i)
        print("Rolling {} {}-sided Die".format(j, i))
        print("Unique Combinations: {}\n".format(result[3]))
        combs[i].append(result[3])
```
