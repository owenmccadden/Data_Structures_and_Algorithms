# Coin Change Problem
- given coins of different denominations and a total amount of money, find the minimum number of coins needed to make up the given amount
- assume each coin has an infinite supply

- solution:
    - find the max denomination that is less than the value at each step
    - add a coin of that denomination to your solution and subtract from total value
    - repeat until the change is made correctly and total value equals 0

```python
import pandas as pd
```

```python
denominations = [1,2,5,10,20,50,100,1000]
print(denominations[::-1])
```

```python
def make_change(value, denoms):
    change = {
        
    }
    for i in denoms:
        change[i] = 0
    coin_count = 0
    
    while value > 0:
        for i in denominations[::-1]:
            if i <= value:
                coin_count += 1
                value -= i
                change[i] += 1
                break
    return coin_count
```

```python
make_change(2048, denominations)
```

```python
x = []
y = []
for i in range(10000):
    x.append(i)
    y.append(make_change(i, denominations))
```

```python
df = pd.DataFrame(y, x)
df.to_csv("coins.csv")
```
